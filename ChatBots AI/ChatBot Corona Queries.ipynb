```python
from newspaper import Article
import random
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import numpy as np
import warnings
```

```python
#Ignore warning messages

warnings.filterwarnings('ignore')
```

```python
#Download packages from NLTK

nltk.download('punkt',quiet=True)
nltk.download('wordnet',quiet=True)
```

```python
#Get the article URL

article = Article('https://en.wikipedia.org/wiki/2019%E2%80%9320_coronavirus_pandemic')
article.download()
article.parse()
article.nlp()
corpus = article.text

#print the corpus text
print(corpus)
```

```python
#tokenization

text = corpus
sent_tokens = nltk.sent_tokenize(text) #Covert the text into a list of sentences

#Print the sentences

print(sent_tokens)
```

```python
# Create a key value pair to remove punctuations

remove_punct_duct = dict( (ord(punt),None) for punt in string.punctuation)

#Print punctuations
print(string.punctuation,'\n\n')

#Print pucntuation count
print(remove_punct_duct)
```

```python
#Return list of lemmatized lower case letter after removing punctuations

def LemNormalize(text):
    return nltk.word_tokenize(text.lower().translate(remove_punct_duct))

#Print the tokenization text
print(LemNormalize(text))
```

```python
#Keyword Matching

#Greeting Inputs

GREETING_INPUTS = ["hi",'hello','hola','greetings','wassup','hey']
GREETING_OUTPUTS = ['howdy','hi','hey',"what's good",'hello','hey there']
```

```python
#Function to return random greeting response

def greetings(sentence):
    for word in sentence.split():
        if word.lower() in GREETING_INPUTS:
            return random.choice(GREETING_OUTPUTS)
```

```python
def response(user_response):    
    #user_response = "what is corona virus"

    user_response = user_response.lower() #make response lower case

    #print(user_response)

    #Chatbot response to empty string
    robo_response = ""

    #Append user response to sentence list(sent_tokens)

    sent_tokens.append(user_response)

    ##print(sent_tokens)

    #Create a TFidfVectorizer object

    TfidfVec = TfidfVectorizer( tokenizer=LemNormalize,stop_words='english')

    #Convert the text to matrix of TF-IDF feature

    tfidf = TfidfVec.fit_transform(sent_tokens)
    ##print(tfidf)

    #Get the measure of similarity (similarity scores)
    vals = cosine_similarity(tfidf[-1],tfidf)
    ###Print the similarity score
    ##print(vals)

    #Get the index of most similat text/sentence to the users response
    idx = vals.argsort()[0][-2]

    #Reduce the dimensionality of vals

    flat = vals.flatten()
    #Sort the list in ascending order

    flat.sort()

    #Get the most similar score
    score = flat[-2]

    #If variable score in 0 then there is no text similar to the user response

    if(score==0):
        robo_response = robo_response+"I apologize, I don't understand"
    else:
        robo_response = robo_response+ sent_tokens[idx]

    #printThe ChatBot response
    ##print(robo_response)
    ##sent_tokens.remove(user_response)

    return robo_response
```

```python

flag = True
print("DOCBot: I'm Doctor Bot or DocBot for short. I'll answer your queries related to Corona, if you want to exit , type Bye!")
while(flag == True):
  user_response = input()
  user_response = user_response.lower()
  if(user_response != 'bye'):
    if(user_response == 'thanks' or user_response =='thank you'):
      flag=False
      print("DOCBot: You are welcome !")
    else:
      if(greetings(user_response) != None):
        print("DOCBot: "+greetings(user_response))
      else:
        print("DOCBot: "+response(user_response))       
  else:
    flag = False
    print("DOCBot: Chat with you later !")
```
